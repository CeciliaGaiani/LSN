### Probabilità: concetti di base

Si definisce **probabilità** la quantificazione dell'aspettativa rispetto all’esito di un esperimento. Si definisce **evento** l’esito di un esperimento che coinvolge casualità.  

Si supponga che un possibile esito di un esperimento sia $A$. Allora, la probabilità che $A$ avvenga è $P(A)$ se, su $N$ esperimenti identici, ci aspettiamo che $N \cdot P(A)$ esiti risultino in $A$. Quando $N\to \infty$, la frazione di esperimenti che risultano in $A$ tenderà proprio a $P(A)$.

Il concetto di spazio campionario è utile per ottenere relazioni tra probabilità per analizzare esperimenti: lo **spazio campionario** di un esperimento è un insieme $\Omega$ tale che ogni esito dell’esperimento corrisponde a uno o più elementi dell’insieme. Un **evento** risulta così un sottoinsieme dello spazio campionario $\Omega$.

Un evento è detto **semplice** se non è ulteriormente decomponibile.

L'**unione** $A_1 \cup A_2$ di due eventi $A_1$ e $A_2$ è l’insieme di tutti gli eventi semplici appartenenti a $A_1$ o $A_2$ o entrambi.

L'**intersezione** $A_1 \cap A_2$ di due eventi $A_1$ e $A_2$ è l’insieme di tutti gli eventi semplici appartenenti sia a $A_1$ sia a $A_2$. Gli eventi $A_1$ e $A_2$ si dicono **mutuamente esclusivi** se $A_1 \cap A_2 = \emptyset$ (insieme vuoto).

$P(A_1\cup A_2)$ è la probabilità che $A_1$ o $A_2$ (o entrambi) si verifichino. Si definisce invece **probabilità congiunta**  $P(A_1\cap A_2)$ la probabilità dell’occorrenza simultanea di $A_1$ e $A_2$. 

La probabilità dell’unione può essere scritta come:
$$
P(A_1\cup A_2)=P(A_1)+P(A_2)-P(A_1\cap A_2)
$$

da cui si osserva che:
- se $A_1$ e $A_2$ sono **mutuamente esclusivi** (nessun evento semplice in comune):
  $$
  P(A_1\cup A_2)=P(A_1)+P(A_2)
  $$

- se $A_1,A_2,\dots,A_m$ sono **mutuamente esclusivi ed esaustivi**:
  $$
  A_1\cup A_2\cup\dots\cup A_m=\Omega
  \quad\text{e}\quad
  \sum_{i=1}^m P(A_i)=1
  $$

Gli eventi $A_1$ e $A_2$ sono **indipendenti** se e solo se:
  $$
  P(A_1\cap A_2)=P(A_1)\cdot P(A_2)
  $$

Si noti che eventi indipendenti non sono necessariamente mutuamente esclusivi.

La **probabilità condizionata** $P(A_2|A_1)$ è la probabilità che l’evento $A_2$ si verifichi dato che $A_1$ si è verificato:

$$
P(A_2|A_1)=\frac{P(A_1\cap A_2)}{P(A_1)}
$$

Dato che $P(A_1\cap A_2)=P(A_2\cap A_1)$, vale anche:

$$
P(A_2|A_1)\,P(A_1)=P(A_1|A_2)\,P(A_2)
$$

Inoltre, se $A_1$ e $A_2$ sono indipendenti:
$$
P(A_2|A_1)=P(A_2)
$$

### Spazio di probabilità e variabili aleatorie
La formalizzazione assiomatica del concetto di probabilità si deve ad Andrey Kolmogorov.
Lo **spazio campionario** è uno spazio di probabilità, cioè uno spazio di misura $(\Omega, \mathcal{F}, \mu)$, definito da:

- un insieme $\Omega$,
- una $\sigma$-algebra $\mathcal{F}$ di sottoinsiemi di $\Omega$,
- e una misura $\mu$ definita su $\mathcal{F}$.

Le proprietà della $\sigma$-algebra sono le seguenti:

- $\Omega \in \mathcal{F}$ e $\emptyset \in \mathcal{F}$ (dove $\emptyset$ è l’insieme vuoto)
- $A\in\mathcal{F} \implies \Omega\setminus A = A'\in\mathcal{F}$ (*complementare*)
- $A,B\in\mathcal{F} \implies A\cap B\in\mathcal{F}$ (*intersezione*)
- $A_i\in\mathcal{F}$, $i=1,2,\dots \implies \bigcup_i A_i\in\mathcal{F}$ (*unione numerabile*)  
  Se inoltre $A_i\cap A_j=\emptyset$ per $i\neq j$, allora:
  $$
  \mu\!\Bigl(\bigcup_i A_i\Bigr)=\sum_i \mu(A_i)
  $$

La collezione di insiemi semplici, insieme all’insieme vuoto e allo spazio intero, genera la $\sigma$-algebra tramite complementari, intersezioni finite e unioni numerabili. La scelta degli *eventi semplici* e della *$\sigma$-algebra* corrispondente è *input* della teoria e dipende dalle informazioni a priori sul problema da trattare.

Un **evento** è un insieme $A$ in $\mathcal{F}$ e la **probabilità** di questo evento è data dalla misura $\mu(A)$. La misura soddisfa la proprietà: $\mu(\Omega) = 1$.

Una **variabile aleatoria f** è una funzione misurabile $f: \Omega \to \mathbb{R}$.

Sfruttando l'assoluta continuità rispetto alla misura di Lebesgue e il teorema di Radon-Nikodym, è possibile riscrivere le misure di probabilità nella forma $d\mu(x) = p(x)\,dx$, dove $p(x)$ è una funzione misurabile chiamata **densità di probabilità (pdf)**, e traslare la struttura probabilistica $(\Omega, \mathcal{F}, \mu)$ in $(\mathbb{R}, \mathcal{B}, p(x)dx)$, dove $\mathcal{B}$ è l'algebra di Borel, cioè la $\sigma$-algebra generata dagli intervalli aperti di $\mathbb{R}$ (o $\mathbb{R}^n$).

Di conseguenza, la probabilità che una variabile casuale assuma un valore all'interno di un sottoinsieme $A \subset \mathbb{R}$ si può esprimere come:

$$
\mathbb{P}(x \in A) = \int_A p(x)\, dx
$$

Si definisce **valore atteso** (o **media**) di una variabile aleatoria $f$ (sfruttando le consoderazioni fatte sopra):

$$
\mathbb{E}[f] := \int_{\Omega} f \, d\mu = \int_{\mathbb{R}} f(x) \, p(x) \, dx
$$

dove:
- $\mu$ è la misura di probabilità su $\Omega$,
- p(x) è la **densità di probabilità** associata a $f$.

La **varianza** della variabile aleatoria $f$ è definita da:

$$
\sigma^2 = \operatorname{Var}(f) = \int_{\mathbb{R}} (f(x) - \mathbb{E}[f])^2 \, p(x) \, dx
$$

La **deviazione standard** (o **scarto quadratico medio**) è la radice quadrata della varianza, e fornisce una stima della tipica distanza dei valori della variabile casuale dal suo valore atteso.

### Matrice di covarianza, correlazione e indipendenza statistica

La **matrice di covarianza** tra un insieme di variabili aleatorie ${x_1, x_2, \dots, x_n}$ è definita come:

$$
\mathrm{cov}[x_i, x_j] \equiv \sigma^2_{ij} = \mathbb{E}[x_i x_j] - \mathbb{E}[x_i] \cdot \mathbb{E}[x_j]
$$

Due variabili aleatorie $x_1$ e $x_2$ si dicono **scorrelate** se la loro covarianza è nulla:

$$
\mathrm{cov}[x_1, x_2] = 0
$$

Due variabili aleatorie $x_1$ e $x_2$ si dicono **statisticamente indipendenti** se la loro densità di probabilità congiunta si fattorizza:

$$
p(x_1, x_2) = p_1(x_1) \cdot p_2(x_2)
$$

In tal caso, vale anche:

$$
\mathbb{E}[x_1 x_2] = \mathbb{E}[x_1] \cdot \mathbb{E}[x_2]
$$

e quindi:

$$
\mathrm{cov}[x_1, x_2] = 0
$$

L'indipendenza implica la scorrelazione, ma non vale il contrario.

### Distribuzioni marginali e condizionate

Si supponga di conoscere la **distribuzione congiunta** di due variabili casuali $p(x_1,x_2)$. Si possono allora definire le loro **distribuzioni marginali**:

$$
p_1(x_1)=\int_{\mathbb{R}} p(x_1,x_2)\,dx_2
\quad\text{e}\quad
p_2(x_2)=\int_{\mathbb{R}} p(x_1,x_2)\,dx_1
$$

La **densità di probabilità condizionata** di $x_1$ dato $x_2$ è definita come:

$$
p(x_1|x_2)=\frac{p(x_1,x_2)}{p_2(x_2)}
$$

(dove si assume $p_2(x_2)>0$). Si può definire simmetricamente $p(x_2|x_1)$.

Per ogni $x_2$ fissato, $p(x_1|x_2)$ è una regolare densità di probabilità in $x_1$ (e analogamente viceversa).

### Teorema del Limite Centrale (CLT)

Si considerino $N$ variabili casuali i.i.d. (indipendenti e identicamente distribuite) $x_1, x_2, \dots, x_N$, ciascuna con:
- media $\mu = \mathbb{E}[x_i]$
- varianza $\sigma^2 = \mathbb{E}[x_i^2] - \mu^2 < \infty$ (**finita**)

La distribuzione di probabilità congiunta è quindi:
$$
p_N(x_1, x_2, \dots, x_N) = \prod_{i=1}^N p(x_i)
$$

Definite la **media campionaria** e la **somma campionaria**:

- Media: $A_N = \frac{1}{N} \sum_{n=1}^N x_n$
- Somma: $S_N = \sum_{n=1}^N x_n = N A_N$

allora valgono le seguenti proprietà per:

1. **Valore atteso:**
   $$
   \mathbb{E}[A_N] = \mu \quad \text{e} \quad \mathbb{E}[S_N] = N \mu
   $$

2. **Varianza:**
   $$
   \text{Var}(A_N) = \frac{\sigma^2}{N} \quad \text{e} \quad \text{Var}(S_N) = N \sigma^2
   $$

3. **Distribuzione asintotica:**

   Quando $N \to \infty$, la distribuzione della media campionaria tende a una gaussiana:
   $$
   p(A_N) \to \frac{1}{\sqrt{2\pi \frac{\sigma^2}{N}}} \exp\left( -\frac{(A_N - \mu)^2}{2 \frac{\sigma^2}{N}} \right)
   $$

   Ovvero:
   $$
   A_N \xrightarrow{d} \mathcal{N}(\mu, \frac{\sigma^2}{N})
   $$

   In alternativa, per la somma:
   $$
   S_N \xrightarrow{d} \mathcal{N}(N\mu, N\sigma^2)
   $$

Dunque, **qualunque sia la distribuzione di partenza di $x_i$, purché $\mu$ e $\sigma^2$ siano finiti, la media campionaria $A_N$ è asintoticamente gaussiana.**

Se si ha a che fare con variabili non indipendenti, il teorema risulta ancora valido a patto che ciascun elemento della sequenza sia correlato con un numero finito di variabili della sequenza.

Si considerino variabili casuali **i.i.d.** $x_i$. $P(S_1)$ coincide con $p(x_i)$. Quando $N$ aumenta, cambia la forma funzionale di $P(S_N)$ e, se le ipotesi del CLT sono soddisfatte, assume la forma gaussiana per valori asintoticamente grandi di $N$. La pdf gaussiana è dunque un **attrattore** nello spazio funzionale delle pdf per tutte le distribuzioni che soddisfano i requisiti del CLT. L’insieme di tali distribuzioni costituisce il **bacino di attrazione** della distribuzione gaussiana.

### Metodo delle medie a blocchi

Si supponga di avere un set di $M$ dati ($x_1, ..., x_M$), dei quali si vuole calcolare valore medio ed incertezza (errore statistico). La formula classica dell'errore:
$$
\sigma_{\langle x \rangle} = \frac{\sigma}{\sqrt{M}}
$$
è valida solo se i dati in questione sono i.i.d. . Se si ha a che fare con dati **correlati**, l'errore calcolato in questa maniera risulterebbe sottostimanto. Per ovviare al problema si utilizza la tecnica del **data blocking**:

1. i dati sono suddivisi in $N$ blocchi da $L = M/N$ elementi ciascuno.
2. si calcola la media in ogni blocco $\bar{x}_j$.
3. si calcola la media globale come **media delle medie di blocco**:
$$
\langle x \rangle = \frac{1}{N} \sum_{j=1}^N \bar{x}_j
$$

4. si stima la **deviazione standard della media di blocco**:
$$
\sigma_{\langle x \rangle}^2 = \frac{1}{N(N - 1)} \sum_{j=1}^N \left( \bar{x}_j - \langle x \rangle \right)^2
$$

5. si stima la **deviazione standard della media** (o errore statistico sulla media stimata) nella media a blocchi:
$$
\sigma_{\langle x \rangle} = \sqrt{ \frac{1}{N(N-1)} \sum_{j=1}^N (\bar{x}_j - \langle x \rangle)^2 }
$$


In questo modo, le medie di blocco risultano **quasi indipendenti**, migliorando la stima dell'errore se i dati sono correlati.

### *Esercizio 01.1.3 - Soluzione proposta*
L'obiettivo dell'esercizio era quello di effettuare un test del $\chi^2$ per valutare se la distribuzione da cui vengono estratti i numeri pseudo-casuali risulti effettivamente uniforme. Successivamente, è stato creato un vettore di $N_{bins}$ componenti, inizialmente impostate a zero, che fungerà da contatore per i numeri appartenenti a ciascun bin. Poi sono state effettuate $N_{\text{throws}} = 10^4$ estrazioni di numeri pseudo-casuali uniformemente distribuiti nell'intervallo $[0,1)$. Per ogni estrazione si è individuato il bin di appartenenza e si è incrementato il contatore relativo.

Una volta ottenuti i conteggi osservati $O_i$ in ciascun bin, si è calcolato il valore del $\chi^2$ mediante la formula:

$$
\chi^2 = \sum_{i=1}^{N_{\text{bins}}} \frac{(O_i - E_i)^2}{E_i}
$$

dove $O_i$ è il numero di conteggi osservato nel bin $i$, $E_i$ è il numero di conteggi atteso, che, ipotizzando una distribuzione uniforme, risulta essere $E_i = \frac{N_{\text{throws}}}{N_{\text{bins}}} = 100$ per ciascun bin.

Il processo è stato ripetuto per 1000 volte, salvando ad ogni iterazione il corrispondente valore di $\chi^2$. Si è scelto di ripetere il test 500 volte (anziché 100) al fine di visulizzare meglio i dati ottenuti.

I risultati del test, ovvero i valori di $\chi^2$ ottenuti, sono riportati nell'istogramma seguente. Si osservi che rispetto alla distribuzione attesa, centrata sul numero di bin $N_{bins} = 100$, quella ottenuta a seguito della simulazione risulta centrata circa intorno a 95. Questo potrebbe sottolineare fluttuazioni statistiche normali o deviazioni del generatore dalla perfetta casualità.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

chi_values = np.loadtxt('Es01.1/chi_square.dat')

df = 100

plt.figure(figsize=(8, 5))
plt.hist(chi_values, bins=20, color='xkcd:cloudy blue', alpha=0.9, density=True, label=r'Istogramma di $\chi^2$')

x = np.linspace(70, 140, 1000)
y = chi2.pdf(x, df)

plt.plot(x, y, color='xkcd:darkish blue', label=r'Distribuzione teorica di $\chi^2$ ' f'(df={df})')

plt.title(r'Test del $\chi^2$', fontsize=14)
plt.xlabel(r'$\chi^2$', fontsize=12)
plt.ylabel("Conteggi", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### *Esercizio 01.2 - Soluzione proposta*
Per la risoluzione di tale esercizio sono stati implementati i generatori di numeri pseudocasuali distributi secondo una distribuzione esponenziale e una distribuzione di Cauchy-Lorentz. Tali metodi sono stati integrati nella classe `Random` del generatore di numeri casuali. Per la costruzione dei generatori è stato usato il metodo dell'inversa della cumulativa.
  
Data una distribuzione di probabilità unidimensionale $p(x)$ definita per $-\infty < x < +\infty$, la **funzione di distribuzione cumulativa** $F(x)$ è definita come:

$$
F(x)=\int_{-\infty}^{x}p(x')\,dx'
$$

La funzione cumulativa soddisfa le seguenti proprietà:
- $\displaystyle \lim_{x\to -\infty}F(x)=0$
- $\displaystyle \lim_{x\to +\infty}F(x)=1$
- $\displaystyle \frac{dF(x)}{dx}=p(x)\geq 0$

La probabilità che $x\in[a,b]$ s i può scrivere come:

$$
P\{x\in[a,b]\}=\int_a^bp(x)\,dx=F(b)-F(a)
$$

Si assume di avere a disposizione una quantità indefinita di variabili pseudo-casuali uniformi. Sia $x$ una variabile casuale con funzione di distribuzione cumulativa $F_x(x)$ e densità di probabilità:

$$
p_x(x)=\frac{dF_x(x)}{dx}
$$

e sia $y=y(x)$ una funzione continua non decrescente/non crescente di $x$.

Si può dimostare che la relazione generale tra le PDF è:

$$
p_y(y)=p_x(x)\left|\frac{dx}{dy}\right| \quad\text{e}\quad p_y(y)\,dy=\pm p_x(x)\,dx
$$

dove il segno $\pm$ dipende dalla monotonia della funzione $y(x)$ ($+$: monotona non decrescente, $-$: monotona non crescente).


Sia $x$ una variabile casuale con distribuzione di probabilità generica $p_x(x)$, e sia $F_x(x)$ la sua funzione cumulativa. Si consideri la variabile $y = F_x(x)$. Dal cambio di variabili nella teoria delle probabilità:

$$
p_y(y) = p_x(x) \cdot \left( \frac{dy}{dx} \right)^{-1} = p_x(x) \cdot \left(\frac{dF_x(x)}{dx}\right)^{-1} = p_x(x) \cdot (p_x(x))^{-1} = \frac{p_x(x)}{p_x(x)} = 1
$$

Poiché $y = F_x(x)$, allora $0 \leq y \leq 1$.

Dunque, la variabile $y = F_x(x)$ è uniformemente distribuita nell’intervallo $[0, 1]$, indipendentemente da $p_x(x)$. Ma allora, se $F_x(x)$ è nota e analiticamente invertibile, generando un numero $y$ uniformemente distribuito $\in [0,1)$:

$$
x = F_x^{-1}(y)
$$

è una variabile casuale con distribuzione $p_x(x)$.

Applicando tale ragionamento alle distribuzioni in questione, sono state calcolate le funzioni cumulative corrispondenti:

- esponenziale: $F(x, \lambda) = 1 - e^{-\lambda x}, \quad x \geq 0, \quad x \geq 0$ 
- Cauchy-Lorentz: $F(x, \lambda, \mu) = \frac{1}{\pi} \arctan\left(\frac{x - \mu}{\gamma}\right) + \frac{1}{2}$

Doopodichè, imponendo:
$$
y_{exp} = F(x_{exp}, \lambda) \quad \text{e \quad} y_{CL} = F(x_{CL}, \lambda, \mu)
$$

si ricavano:

- generatore di numeri distribuiti esponenzialmente ($\lambda$): $x_{exp} = -\frac{1}{\lambda} \ln(1 - y_{exp})$
- generatore di numeri distribuiti secondo Cauchy-Lorentz ($\lambda, \mu$): $x_{CL} = \mu + \gamma \cdot \tan\left[ \pi \left( y_{CL} - \frac{1}{2} \right) \right]$

dove $y_{exp}$ e $y_{CL}$ sono uniformemente distribuiti in $[0, 1)$.
 
Nella seconda parte dell'esercizio sono stati utilizzati i generatori implementati per verificare il CLT. Si calcola $M=10^4$ volte la quantità:

$$
S_N = \frac{1}{N}\sum_{i=1}^N x_i \quad \text{con $N=1, 2, 10, 100$}
$$

con $x_i$ estratti da distribuzioni uniforme, esponenziale ($\lambda$ = 1) e di Cauchy-Lorentz ($\mu = 0$ e $\lambda = 1$).

Per le distribuzioni uniforme ed esponenziale, la distribuzione della variabile $S_N$ converge a una distribuzione gaussiana, con media pari al valore atteso e varianza pari alla varianza della distribuzione di partenza divisa per $N$, in accordo con il CLT.

Nel caso della distribuzione di Cauchy–Lorentz la distribuzione limite non è una Gaussiana. Questo accade perché la Cauchy–Lorentz non soddisfa le ipotesi del Teorema del Limite Centrale, in quanto presenta secondo momento infinito.
La distribuzione di Cauchy è un esempio di una classe di distribuzioni che possiedono code lunghe a legge di potenza inversa:

$$
p(x)\sim \frac{1}{|x|^{1+\alpha}}, \quad 0<\alpha<2, \quad x\to\infty
$$

le quali precludono la convergenza alla gaussiana per $N\to\infty$. Il caso Cauchy–Lorentz rientra nella generalizzazione del Teorema del Limite Centrale, in cui la convergenza avviene a una distribuzione stabile di Lévy (in questo caso, la distribuzione di Cauchy-Lorentz stessa).

### *Esercizio 01.3 - Soluzione proposta*
Il metodo consiste di Buffon per la misura di $\pi$ consiste nel lanciare un ago di lunghezza $L$ su una superficie su cui sono tracciate linee orizzontali equidistanti, con distanza $d$, tale che $d>L$.

Detta $P$ la probabilità che l'ago intersechi una delle righe, si può dimostrare che $P = \frac{2L}{\pi d}$, da cui si ricava una formula per stimare $\pi$:

$$
\pi = \frac{2L N_{thr}}{ N_{hit} d}
$$

dove $N_{thr}$ è il numero totale di lanci dell’ago e $N_{hit}$ è il numero di volte in cui l’ago ha effettivamente intersecato una linea (il rapporto $N_{hit}$/$N_{thr}$ è usato per fornire una stima di $P$).

Per rendere la stima statisticamente più robusta, si suddivide l’esperimento in $N = 100$ blocchi da $L = N_{thr} / N = 5000$ lanci ciascuno, con un totale di $N_{thr} = 5 \cdot 10^5$ lanci. $N_{thr}$ è stato scelto più alto rispetto alla richiesta dell'esercizio in modo da garantire una migliore stima del valore di $\pi$. Ogni blocco fornisce una stima indipendente di $\pi$; l’insieme di queste viene poi utilizzato per calcolare medie progressive e incertezze statistiche mediante il metodo del data blocking.

Per ciascun lancio, vengono estratti due valori casuali:

- La distanza $r$ tra il centro dell’ago e la linea più vicina, campionata uniformemente in $[0, d/2)$.
- L’angolo $\theta$ che l’ago forma con la direzione delle linee (supposta coincidente con l’asse $x$), campionato uniformemente in $[0, \pi)$.

L'estrazione dell'angolo utilizza il metodo del rigetto. Si estraggono due numeri pseudocasuali $x, y$ uniformemente distribuiti in $[-1, 1]$ (in un quadrato di lato unitario). Si accettano solo le coppie tali che $x^2 + y^2 \leq 1$, ovvero solo i punti all’interno del disco unitario (inscritto nel rettangolo). Questo garantisce un campionamento uniforme all’interno del cerchio. Si calcola l’angolo $\theta$ formato dal vettore $(x, y)$ con l’asse delle ascisse come:
$$
\theta = \arccos\left(\frac{x}{\sqrt{x^2 + y^2}}\right)
$$

Il valore di $\theta$ così ottenuto appartiene all'intervallo $[0, \pi]$. Poiché la funzione $\sin(\theta)$ è simmetrica rispetto a $\pi/2$ nell'intervallo $[0, \pi]$, e poichè l’ago è simmetrico rispetto al proprio centro, la distribuzione angolare ottenuta è corretta: gli angoli $\theta$ e $\pi - \theta$ portano allo stesso risultato fisico. Dunque, non è necessario un campionamento in $[0, 2\pi)$.

A questo punto si calcola la proiezione `proj` della semi-lunghezza dell'ago lungo l'asse y, come $\frac{L}{2}sen(\theta)$. Si verifica se la proiezione dell’ago risulti maggiore della distanza del centro dell'ago dalla riga più vicina (`proj` ≥ `pos`). In tal caso significa che l'ago interseca effettivamente la linea, dunque si incrementa il contatore $N_{hit}$.

Il grafico sottostante mostra l’andamento delle medie progressive, con le relative incertezze, in funzione del numero di blocchi considerati. Si osserva una buona compatibilità con il valore atteso di $\pi$.
